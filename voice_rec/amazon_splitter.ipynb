```python


# Synthesize the sample text, saving it in an MP3 audio file
response = polly.synthesize_speech(VoiceId='Joanna',
                                          OutputFormat='json',
                                          Text='This is sample text to synthesize. Go and fuck yorself volchara ebanaya', SpeechMarkTypes = ["word","sentence"])
with open('output.json', 'wb') as f:
    #json.dump(spoken_text['AudioStream'].read(), f, indent = 4)
    f.write(response['AudioStream'].read())
    
print(response)

response = polly.synthesize_speech(VoiceId='Joanna',
                                          OutputFormat='mp3',
                                          Text='This is sample text to synthesize. Go and fuck yorself volchara ebanaya')
with open('output.mp3', 'wb') as f:
    #json.dump(spoken_text['AudioStream'].read(), f, indent = 4)
    f.write(response['AudioStream'].read())
    
print(response)
```

```python



spoken_text = polly.synthesize_speech(Text='I am Emma. You know me from Ivona, but now I come via Polly.',
                                      OutputFormat='mp3',
                                      VoiceId='Emma')

with open('output.mp3', 'wb') as f:
    f.write(spoken_text['AudioStream'].read())
    f.close()

mixer.init()
mixer.music.load('output.mp3')
mixer.music.play()

while mixer.music.get_busy() == True:
    pass

mixer.quit()
os.remove('output.mp3')
```
