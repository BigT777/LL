```python
import re
from tqdm import tqdm
from string import punctuation
full_punctuation = punctuation + "–" + "," + "»" + "«" + "…" +'’'
```

```python
from nltk.corpus import stopwords
stopWords = set(stopwords.words('english'))
```

```python
full_punctuation
```

```python
cleaned_bigramm = []
filtered_count = 0
filtered_stuff = []
with open("/Users/nigula/Desktop/long_term_storage/ngramms_final/final_bigramms.txt", 'r') as f:
    for line in tqdm(f.readlines()):
        look_for_digit = re.findall('[1-9]',line)
        if 'http' in line or len(look_for_digit) > 0:
            filtered_stuff.append(line)
            filtered_count += 1
        else:
            cleaned_bigramm.append(line[:-1])
print(filtered_count)
print(filtered_stuff[:20])
```

```python
len(cleaned_bigramm)
```

```python
with open("final_bigramm_cleaned.txt","w") as f:
    for bigr in cleaned_bigramm:
        f.write(bigr + '\n')
```

```python
final_trigramms = []
filtered_count = 0
filtered_stuff = []
with open("/Users/nigula/Desktop/long_term_storage/ngramms_final/final_trigramms.txt", 'r') as f:
    for line in tqdm(f.readlines()):
        line_list = line.split()
        #print(line_list)
        look_for_digit = re.findall('[1-9]',line)
        look_for_punct = re.findall('[!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~–,»«…’]',line)
        
        if 'http' in line or len(look_for_digit) > 0 or len(look_for_punct)>0 or line_list[-1] in stopWords:
            filtered_stuff.append(line)
            filtered_count += 1
        else:
            final_trigramms.append(line[:-1])
print(filtered_count)
print(filtered_stuff[:20])
```
