```python



def translate_word_branch(word_branch_sences_list, word_id, eng_word, lemma_id,
                         word_id_list,word_value,lemma_id_list,definition,local_word,examples,definition_local,
                             examples_local,word_rating):
    for lemma_sence_list_element in word_branch_sences_list:
        word_id_list.append(word_id)
        word_value.append(eng_word)
        lemma_id_list.append(lemma_id)
        if 'definition' in lemma_sence_list_element:
            definition.append(lemma_sence_list_element['definition'])
            definition_local.append(lemma_sence_list_element['definition_local'])
        else:
            definition.append(None)
            definition_local.append(None)
        examples.append(lemma_sence_list_element['examples'])
        examples_local.append(lemma_sence_list_element['examples_local'])
        local_word.append(lemma_sence_list_element['local_word'])
        word_rating.append(lemma_sence_list_element['word_rating'])
        
def json2df(new_dict):
    word_id_list = []
    word_value = []
    lemma_id_list = []
    definition = []
    local_word =[]
    examples = []
    definition_local = []
    examples_local = []
    word_rating = []
    for eng_word in new_dict.keys():
        #print(eng_word)
        translate_word_branch(new_dict[eng_word]['lemma_branch'],new_dict[eng_word]['word_id'],eng_word, new_dict[eng_word]['word_id'],
                             word_id_list,word_value,lemma_id_list,definition,local_word,examples,definition_local,
                             examples_local,word_rating )
        for wordform in new_dict[eng_word]['wordforms_branch'].keys(): 
            translate_word_branch(new_dict[eng_word]['wordforms_branch'][wordform]['wordsenses'],
                                 new_dict[eng_word]['wordforms_branch'][wordform]['word_id'],
                                 wordform, new_dict[eng_word]['word_id'],
                             word_id_list,word_value,lemma_id_list,definition,local_word,examples,definition_local,
                             examples_local,word_rating )
        #print("======\n")  
    #return new_dict
    
    df = pd.DataFrame(list(zip(word_id_list,
    word_value,
    lemma_id_list, 
    definition,
    local_word,
    examples,
    definition_local,
    examples_local,
    word_rating)), 
    columns =['word_id', 'word_value', 'lemma_id','definition','local_word','examples','definition_local',
                             'examples_local','word_rating'])
    return df

#df = json2df(big_dict_with_almost_all_features)
```
