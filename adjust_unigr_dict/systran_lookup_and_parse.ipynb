```python
def parse_json(path, word, local_word,examples,examples_local, examples_and_definitions,  print_output = False):
    
    with open(path, "r") as f: 
        data = json.load(f)
        for output in data['outputs']:
            for match in output['output']['matches']:
                word_current = match['source']['lemma']
                #pos = match['pos']
                print("POS DETECTED",match)
                for target in match['targets']:
                    if print_output:
                        print(target)
                    word.append(word_current)
                    local_word.append(target['lemma'])
                    ex_en = ''
                    ex_rus = ''
                    if len(target['expressions']) == 0:
                        ex_en = "empty_cell"
                        ex_rus = "empty_cell"
                    else:
                        for example in target['expressions']:
                            ex_en += example['source'] + '|'
                            ex_rus += example['target'] + ' | '
                            if print_output: print(example)
                    examples.append(ex_en)
                    examples_local.append(ex_rus)
                    if len(target['invmeanings']) == 0 or target['invmeanings'][0]=="": 
                        pass
                    else:
                        if ex_en == "empty_cell":
                            ex_en = ''
                        for definit in target['invmeanings']:
                            ex_en += definit + ' | '
                    examples_and_definitions.append(ex_en)

                    if print_output:print()
   
    
def parse_from_systran_lookup_vs_wordforms(directory):
    word = []
    local_word = []
    examples = []
    examples_local = []
    definition_reference = []
    words = pd.read_csv("test_words.csv", header = None)
    for word_i in words[0].items():
        print (word_i[1], "========")
        file = word_i[1] + ".json"
        parse_json(os.path.join(directory,file),word, local_word,examples,examples_local,definition_reference, print_output=True)
        
    data = pd.DataFrame(list(zip(word,local_word,examples, examples_local, definition_reference)),columns =['word', 'local_word', 'examples','local_examples', 'examples_and_definitions'])
    return  data
df = parse_from_systran_lookup_vs_wordforms("/Users/nigula/LL/adjust_unigr_dict/lookup/systran_en_ru")
```
