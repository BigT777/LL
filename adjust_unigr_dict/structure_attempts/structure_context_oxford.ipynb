# CONTEXT (SENCE-WORD AND CORRESPONDING EXAMPLES)

```python
def get_definitions(collect_wordsence_examples_dict, word, from_lang, to_lang):
    collect_wordsence_examples_dict[word] = {}
    url = "https://context.reverso.net/перевод/" + from_lang + "-" + to_lang + "/" + word 
    print(url)
    login = {'inUserName': 'n.babakov@lingualeo.com', 'inUserPass': '33vec33'}
    header_main = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    response = requests.get(url, headers=header_main, data = login)
    response.encoding = 'utf-8' 
    soup = bs(response.text, 'html.parser')
    first_string_passed = False
    translations_variants = []
    for link in soup.find_all("a", attrs={"class" : "translation"}):
        if first_string_passed == True:
            sence_word = link.text.strip()
            print(sence_word)
            translations_variants.append(sence_word)
        first_string_passed = True
    time.sleep(2)
    for transl_word in translations_variants:
        time.sleep(random.uniform(0.4,2.9))
        url_example = "https://context.reverso.net/перевод/" + to_lang + "-" + from_lang + "/" + transl_word
        response_ex = requests.get(url_example, headers=header_main)
        soup_sub = bs(response_ex.text, 'html.parser')
        print("========================================")
        print("\nEXAMPLES FOR SENCE ",transl_word,"\n" )
        example_occured = False
        for l in soup_sub.find_all('span', attrs = {'class':'text', 'lang':'en'}):
            example = l.text.strip()
            #print(example, word in example, type(l.text.strip()), word)
            regex = "\W" + word + "\w{0,2}\W"
            find = re.findall(regex, example)
            if len(find) > 0 :
                if example_occured == False:
                    collect_wordsence_examples_dict[word][transl_word] = []
                    collect_wordsence_examples_dict[word][transl_word].append(example)
                    example_occured = True
                else:
                    collect_wordsence_examples_dict[word][transl_word].append(example)
                print(example)
        
 

```

# Sences from wordnet

```python
words_for_eval = ['go','report','drive','look','feel','people','learn','table','book']
sences_dict = {}
def get_definitionas_and_examples(word, sence_dictionary):
    sence_dictionary[word] = []
    for i, synset in enumerate(wn.synsets(word)):
        currecnt_definition_json = {'definition':synset.definition()}
        currecnt_definition_json['example'] = []
        for example in synset.examples():
            currecnt_definition_json['example'].append(example)
        sence_dictionary[word].append(currecnt_definition_json)
```

```python
def get_sence_examples_list(word,miltisence_json,  print_output = True):
    miltisence_json[word] = []
    endpoint = "entries"
    language_code = "en-us"
    url = "https://od-api.oxforddictionaries.com/api/v2/" + endpoint + "/" + language_code + "/" + word.lower()
    r = requests.get(url, headers = {"app_id": app_id, "app_key": app_key})
    results_list = r.json()['results']
    for result in results_list:
        if print_output == True:
            print("NEW result",word)
        for lex_entr in result['lexicalEntries']:
            if print_output == True:
                print("==NEW lexentry")
            for entry in lex_entr['entries']:
                if print_output == True:
                    print("======NEW entry")
                for sence in entry['senses']:
                    current_sence = {}
                    if print_output == True:
                        print("==========NEW sence")
                        print(sence['definitions'][0])
                        if 'examples' in sence:
                            for ex in sence['examples']:
                                print(ex['text'])
                        print()

                    current_sence['definition'] = sence['definitions']
                    if 'examples' in sence:
                        current_sence['examples'] = []
                        for ex in sence['examples']:
                            current_sence['examples'].append(ex['text'])
                    miltisence_json[word].append(current_sence)

                    if 'subsenses' in sence:
                        if print_output == True:
                            for subsence in sence['subsenses']:

                                print("==============NEW subsence")
                                print(subsence['definitions'][0])
                                if 'examples' in subsence:
                                    for ex in subsence['examples']:
                                        print(ex['text'])
                                print()

                        for subsence in sence['subsenses']:
                            current_subsence = {}
                            current_subsence['definition'] = subsence['definitions']
                            if 'examples' in subsence:
                                current_subsence['examples'] = []
                                for ex in subsence['examples']:
                                    current_subsence['examples'].append(ex['text'])
                            miltisence_json[word].append(current_subsence)
#multisence_dict_d = {}
                        
#get_sence_examples_list('go',multisence_dict_d)
```
