```python
data = {'name_surname':[],'comment_text':[],'time':[] }
for id in range (148501, 149426,100):
    print ("start from id = ", id)
    res = api.wall.getComments(owner_id = '-15787787',post_id = '148500',start_comment_id = id, count = 100)
    print("comments in this instance", len(res['items']))
    for post in res['items']:
        account_info = api.users.get(user_ids = post['from_id'])
        ts = int(post['date'])
        name_surname = account_info[0]['first_name'] + ' ' + account_info[0]['last_name'] 
        comment_text = post['text']
        time = datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
        print(name_surname,comment_text,time)
        data['name_surname'].append(name_surname)
        data['comment_text'].append(comment_text)
        data['time'].append(time)
    print("totally collected", len(data['name_surname']),len(data['comment_text']), len(data['time']))
```
